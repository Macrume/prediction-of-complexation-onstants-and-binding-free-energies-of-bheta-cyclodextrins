```python
import pandas as pd
import sys
from pathlib import Path

project_root = Path.cwd()
if (project_root / "predicting-cyclodextrin-bioavailability").is_dir():
    project_root = project_root / "predicting-cyclodextrin-bioavailability"

sys.path.append(str(project_root))

from scripts.utils import env_variables as env


NEW_FILE = env.NEW_CD_DATA
OLD_FILE = env.CD_ENRICHED_DATA
OUT_FILE = env.MERGED_CD_DATA


new = pd.read_csv(NEW_FILE, sep=";", decimal=",", low_memory=False)
old = pd.read_csv(OLD_FILE, sep=",", decimal=".", low_memory=False)



for col in ["CID_Host", "CID_Guest"]:
    new[col] = pd.to_numeric(new[col], errors="coerce").astype("Int64")
    old[col] = pd.to_numeric(old[col], errors="coerce").astype("Int64")


for col in ["pH", "T"]:
    new[col] = pd.to_numeric(new[col], downcast="integer", errors="coerce")
    old[col] = pd.to_numeric(old[col], downcast="integer", errors="coerce")


key_cols = ["Host", "CID_Host", "Guest", "CID_Guest", "pH", "T", "K"]


dup_new = new[new.duplicated(key_cols, keep=False)]
dup_old = old[old.duplicated(key_cols, keep=False)]

print("Duplikaty w NEW:", dup_new.shape[0])
print("Duplikaty w OLD:", dup_old.shape[0])

# podejrzyj kilka przykładowych powtórek
if not dup_new.empty:
    print("\nPrzykład z NEW:\n", dup_new[key_cols + ["DeltaG"]].head())
if not dup_old.empty:
    print("\nPrzykład z OLD:\n", dup_old[key_cols + ["DeltaG"]].head())
```

```python


new = new[
    (new["Host"] == "beta-cyclodextrin") &
    (new["pH"] == 7) &
    (new["T"] == 298)
].copy()


merged = pd.merge(
    new, old,
    on=key_cols,
    how="left",
    suffixes=("", "_old"),
    indicator=True          # pozwala zobaczyć statystyki dopasowań
)

print(merged["_merge"].value_counts())  # podgląd dopasowań


for col in new.columns:
    if col in key_cols:          
        continue
    old_col = f"{col}_old"
    if old_col in merged.columns:
        merged[col] = merged[col].combine_first(merged[old_col])
        merged.drop(columns=old_col, inplace=True)

merged.drop(columns="_merge", inplace=True)
print(merged.isna().sum().sort_values().tail(20))



merged.to_csv(OUT_FILE, sep=";", decimal=",", index=False)

print("Scalona baza zapisana do:", OUT_FILE)
print("Rozmiar końcowy:", merged.shape)
```
